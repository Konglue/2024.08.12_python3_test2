```python
import numpy as np

def calculate_equilibrium_constant(a_concentration, b_concentration, c_concentration, d_concentration):
    """
    Calculate the equilibrium constant for a given chemical reaction.

    Parameters:
    a_concentration (float): Concentration of A
    b_concentration (float): Concentration of B
    c_concentration (float): Concentration of C
    d_concentration (float): Concentration of D

    Returns:
    float: The equilibrium constant.
    """
    # Coefficients for the reaction (example: 2A + B ↔ C + 2D)
    a_coeff = 2
    b_coeff = 1
    c_coeff = 1
    d_coeff = 2

    # Calculate the concentrations of products and reactants
    product_concentration = (c_concentration ** c_coeff) * (d_concentration ** d_coeff)
    reactant_concentration = (a_concentration ** a_coeff) * (b_concentration ** b_coeff)

    # Calculate equilibrium constant
    if reactant_concentration == 0:
        raise ValueError("Reactant concentration cannot be zero.")

    K = product_concentration / reactant_concentration
    return K

# Input concentrations
a_concentration = float(input("Enter concentration of A: "))
b_concentration = float(input("Enter concentration of B: "))
c_concentration = float(input("Enter concentration of C: "))
d_concentration = float(input("Enter concentration of D: "))

try:
    # Calculate equilibrium constant
    K = calculate_equilibrium_constant(a_concentration, b_concentration, c_concentration, d_concentration)
    print(f"The equilibrium constant K is: {K:.2f}")
except ValueError as e:
    print(f"Error: {e}")
```
